# Analysis 1

- Relation between fertility rate and literacy in women in densely populated countries of Asia and Africa.


```python
# import pandas and read csv 
import pandas as pd
import numpy as np

worldBankData = pd.read_csv("/Users/vega/Downloads/WDI_csv/WDI_Data.csv")
```

```python
#Keep only data for countries which has value in atleast one year
yearColumns = [col for col in worldBankData.columns if col.isdigit()]
worldBankData = worldBankData.dropna(subset=[yearColumns], how='all')
#Drop columns not relevant to search
worldBankData = worldBankData.drop(['Country Code', 'Indicator Code'], axis=1)
```

```python
# Get values for indicators on Fertility rate and literacy for women
indc_list = ['Fertility rate, total (births per woman)','Literacy rate, adult female (% of females ages 15 and above)']
worldBankData = worldBankData[worldBankData['Indicator Name'].isin(indc_list)]
worldBankData = worldBankData.set_index(['Country Name','Indicator Name']).stack().unstack(1)
```

```python
#reset index and set the column names as required
worldBankData = worldBankData.reset_index()
new_columns = worldBankData.columns.values;
worldBankData.columns = ['Country Name', 'Year','Fertility rate, total (births per woman)','Literacy rate, adult female (% of females ages 15 and above)']
worldBankData = worldBankData.dropna()
```

```python
#List of Asian and African countries
worldBankData_Asia = worldBankData[worldBankData['Country Name'].isin(['Philippines','China','Indonesia','Pakistan'])]
worldBankData_Africa = worldBankData[worldBankData['Country Name'].isin(['South Africa','Morocco','Tunisia','Uganda'])]

```

```python
#Create seaborn plot for Analysis of Literacy rate of Women vs Fertility rate in Asia

%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt

g = sns.FacetGrid(worldBankData_Asia, col='Country Name', sharey=False, sharex=False, margin_titles=True,size=4, aspect=1,col_wrap=2)
g.map(plt.scatter, "Literacy rate, adult female (% of females ages 15 and above)", "Fertility rate, total (births per woman)")
g.map(plt.plot, "Literacy rate, adult female (% of females ages 15 and above)", "Fertility rate, total (births per woman)")
g.fig.suptitle("Analysis of Literacy rate of Women vs Fertility rate in Asia", size=16)
g.fig.subplots_adjust(top=.88)
g.set_xlabels('Literacy rate(% of adult female 15 yrs & above)')
g.savefig("ana_1.1.png")
```

```python
#Create seaborn plot for Analysis of Literacy rate of Women vs Fertility rate in Africa
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt

g = sns.FacetGrid(worldBankData_Africa, col='Country Name', sharey=False, sharex=False, margin_titles=True,size=4, aspect=1,col_wrap=2)
g.map(plt.scatter, "Literacy rate, adult female (% of females ages 15 and above)", "Fertility rate, total (births per woman)")
g.map(plt.plot, "Literacy rate, adult female (% of females ages 15 and above)", "Fertility rate, total (births per woman)")

g.fig.suptitle("Analysis of Literacy rate of Women vs Fertility rate in Africa", size=16)
g.fig.subplots_adjust(top=.88)
g.set_xlabels('Literacy rate(% of adult female 15 yrs & above)')
g.savefig("ana_1.2.png")
```

# Conclusion
- Based on the plots above,we can conclude that as the literacy rate in women increases,the fertility rate decreases sharply in densely populated countries of Asia and Africa.
